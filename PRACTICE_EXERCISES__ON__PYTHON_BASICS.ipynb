#1.Write a Python program which accepts the radius of a circle from the user and compute the area.


```python
import math

# Accept radius from the user
radius = float(input("Enter the radius of the circle: "))

# Calculate area using the formula: area = π × r²
area = math.pi * radius ** 2

# Display the result
print("Area of the circle =", round(area, 2))
```

#3.Write a Python program to get the largest number from a list.Sample input: 12, 3, 47, 10 Sample output: 47 is the largest number.

```python
# Define the list of numbers
numbers = [12, 3, 47, 10]

# Find the largest number using max()
largest = max(numbers)

# Display the result
print(f"{largest} is the largest number.")
```

#4.Given a range of first 10 numbers, write a Python program to iterate from start number to the end number and print the sum of the current number and previous number. Sample input: 1….10 Sample output: Current Number 1 Previous Number 0 Sum: 1 Current Number 2 Previous Number 1 Sum: 3 Current Number 3 Previous Number 2 Sum: 5… Current Number 10 Previous Number 9 Sum: 19

```python
previous = 0

for current in range(1, 11):
    total = current + previous
    print(f"Current Number {current} Previous Number {previous} Sum: {total}")
    previous = current
```

#5.Write a Python program to print only those numbers which are divisible of 5. Sample input: 10, 20,33,46,55 Sample output: 10, 20, 55

```python
# Sample input list
numbers = [10, 20, 33, 46, 55]

# Filter and print numbers divisible by 5
divisible_by_5 = [str(num) for num in numbers if num % 5 == 0]
print(", ".join(divisible_by_5))
```

#6.Write a Python program to check whether a number is prime or not. Sample input: Enter the number: 3 Sample output: 3 is a prime number

```python
# Accept input from the user
num = int(input("Enter the number: "))

# Check if the number is prime
if num <= 1:
    print(f"{num} is not a prime number")
else:
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(f"{num} is a prime number")
    else:
        print(f"{num} is not a prime number")
```

#7.Write a Python program to reverse a list using for loop. Sample input: 10,40,30,70 Sample output: 70,30,40,10


```python
# Original list
numbers = [10, 40, 30, 70]

# Create an empty list to store reversed elements
reversed_list = []

# Iterate through the original list in reverse order
for i in range(len(numbers) - 1, -1, -1):
    reversed_list.append(numbers[i])

# Display the reversed list
print(", ".join(str(num) for num in reversed_list))
```

#9.Write a Python program to find the maximum of three numbers Sample input: 34,12,7 Sample output: 34


```python
# Accept three numbers from the user
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

# Find the maximum using if-else
if a >= b and a >= c:
    maximum = a
elif b >= a and b >= c:
    maximum = b
else:
    maximum = c

# Display the result
print(f"{maximum} is the largest number.")
```

#10.Write a Python program to construct the following pattern, using a nested for loop.
*
**
***
****
*****
****
***
**
*

```python
# Number of rows for the top half
rows = 5

# Top half of the pattern
for i in range(1, rows + 1):
    for j in range(i):
        print('*', end='')
    print()

# Bottom half of the pattern
for i in range(rows - 1, 0, -1):
    for j in range(i):
        print('*', end='')
    print()
```
